```python
import pandas as pd
from pandas import DataFrame, Series
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import os

path='/home/std/Downloads/python/test_result/piplineV2_overfit/'
out=dict()
for file in os.listdir(path):
    data=pd.read_csv(path+file)
    for i in range(len(data)):
        if data.iloc[i]['model_tag'] not in out.keys():
            out[data.iloc[i]['model_tag']] = 1
        else:
            out[data.iloc[i]['model_tag']] +=1
print(out)

g=pd.DataFrame(list(out.items()),columns = ['job','value']) 
print(g.head())

plt.figure(figsize=(15,30))
sns.barplot(x='value',y='job',data=g)
```
