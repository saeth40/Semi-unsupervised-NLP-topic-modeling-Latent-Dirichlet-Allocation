# load data

```python
import pandas as pd
from pandas import DataFrame, Series
import numpy as np

def modelSelection(name,error,rep):
  tagName=pd.read_csv(name)
  tagError=pd.read_csv(error)
  tagRep=pd.read_csv(rep)
  yError=tagError.drop(['Unnamed: 0','model','n_below','n_topic', 'alpha', 'eta'], axis=1)
  frames=[]
  for i in list(yError.columns):
      kk=tagError[tagError[i]==min(tagError[i])][['alpha', 'eta', 'n_topic']]
      kk.reset_index(inplace=True)
      #print(i, min(tagError[i]))
      #print(kk)
      for j in range(len(kk)):
        out=pd.DataFrame()
        alp=str(kk.iloc[j]['alpha'])
        beta=str(kk.iloc[j]['eta'])
        top=str(kk.iloc[j]['n_topic'])
        #print(alp, beta, top)
        #[below:topic:alpha:eta]_15:100:0.005:None.sav
        out['job']=pd.Series([i[6:]])
        out['alpha']=pd.Series([alp])
        out['beta']=pd.Series([beta])
        out['n_topic']=pd.Series([top])
        for h in tagRep[tagRep['model']== '[below:topic:alpha:eta]_15:'+ top+':'+alp+':'+beta+'.sav'][i[6:]]:
          out['Represent_topic']=pd.Series([h])
        out['acc_first']= (1-min(tagError[i]))*100
        k=tagName[tagName['tag_topic']==i[6:]]['[below:topic:alpha:eta]_15:'+top+':'+alp+':'+beta+'.sav'].value_counts()
        x=list(k.index)
        y=[(h*100)/sum(list(k)) for h in list(k)]
        for t in range(1,4):
          if t<len(x):
            out[str(t+1)+'_guest']=pd.Series([x[t]])
            out['acc_'+str(t+1)+'_guest']=pd.Series([y[t]])
          else:
            out[str(t+1)+'_guest']=pd.Series(['-'])
            out['acc_'+str(t+1)+'_guest']=pd.Series(['-'])
        frames.append(out)
  allData=pd.concat(frames)
  print('finish')
  return allData
```
