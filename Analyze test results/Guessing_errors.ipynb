```python
import pandas as pd
from pandas import DataFrame, Series
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

#fault prediction function
def fp(csv):
  data=pd.read_csv(csv)
  jobs=list(data['tag_topic'].unique())
  model=list(data.drop(['Unnamed: 0','tag_topic'],axis=1).columns)
  for j in jobs:
    temp=dict()
    for m in model:
      k=data[data['tag_topic']==j][m].value_counts()
      x=list(k.index)
      y=[((h*100)/sum(list(k)))/len(model) for h in list(k)]
      for i in range(len(x)):
        if x[i] not in temp.keys():
          temp[x[i]]=y[i]
        else:
          temp[x[i]]+=y[i]
    print('#####'+j+'\n')
    for i in dict(sorted(temp.items(), key=lambda item: item[1],reverse=True)).items():
      print(i)
    print('\n')
```
