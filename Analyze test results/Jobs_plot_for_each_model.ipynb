```python
job_label=list(raw_data['tag_topic'].unique())
target = ['stem', 'therapist', 'coordinator high skill', 'buesiness develop', 'coordinator low skill', 'finance manager',
          'investor + trader + risk management', 'product', 'supply and distribution manager', 'event organizer', 
          'teacher art', 'teacher academics', 'flim', 'gardener', 'craftman', 'teacher music', 'teacher sport',
          'policy and planing manager', 'entertainer', 'business service and admin manager', 'health service manager',
          'education manager', 'tour guider', 'librarian', 'artist', 'drama writer', 'r and d manager', 'photographer',
          'advertising and public relation manager', 'interior archetecture', 'writing creator', 'other managers']
for i in target:
    job_label.pop(job_label.index(i))
job_label
```

```python
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

raw_column=list(raw_data.columns)
raw_column.remove('tag_topic')
select_column=raw_column
#job distribution for 1 model

warnings.filterwarnings('ignore')

for model in select_column[:30]:
  plt.figure(figsize=(15,10))
  for job in job_label:
    sns.distplot(raw_data[raw_data['tag_topic']==job][model], label=job, bins=1)
    plt.legend()
```
